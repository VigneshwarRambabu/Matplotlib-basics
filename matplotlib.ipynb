```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
import numpy as np
```

```python
x = np.linspace(0,25,5)
y = x*2
z= y+2
```

```python
x
```

```python
y
```

```python
z
```

```python
#Functional Plot
```

```python
plt.plot(x,y)
plt.xlabel('X value')
plt.ylabel("Y value")
plt.title("TITLE")
plt.show()
```

```python
from matplotlib import style 
```

```python

style.use("ggplot")
plt.plot(x,y,'b', label ="first line", linewidth = 1)
plt.plot(x,z,'g', label ="first line", linewidth = 3)
plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
#SUBPLOT
```

```python
plt.subplot(1,2,1)
plt.plot(z, x,'-*')
plt.subplot(1,2,2)
plt.plot(y, x,'o')
```

```python
# object oriented
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.12,0.84,0.85])
axes1.plot(x,y,'o',label = 'outer plot ')
plt.legend()
axes1.set_xlabel("Row value")
axes1.set_ylabel("Column value")
axes1.set_title("OO plotting ")
axes2 = fig.add_axes([0.5,0.4,0.4,.5])
axes2.plot(y,x,'v',label = 'inner plot ')
plt.grid(True, color= 'k')
plt.legend()
```

```python
#SUBPLOT
```

```python
fig, axes = plt.subplots(nrows=3,ncols=2)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'-c')
axes[1].plot(z,x,'-*')
```

```python
#figureszie(figure width and height in inches) and DPI(dots per inch)
```

```python
fig,axes = plt.subplots(nrows= 2, ncols=2, figsize=(4,4), dpi=100)
plt.tight_layout()
fig1 = axes[0][1].plot(x,y)
```
